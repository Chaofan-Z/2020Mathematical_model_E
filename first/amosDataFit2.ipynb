```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dataorigin = pd.read_csv('./goodData/amos20200313.csv')
data = pd.read_csv('./goodData/amos0313wyl.csv')
def cut(x):
    if x >= 7000:
        return 7000
    elif x >= 5000:
        return 5000
    else:
        return x
data['MOR_1A_cut'] = data['MOR_1A'].apply(cut)
data.describe()
```

```python
color = ["green", "blue", "red", "black"]

temp = []
for i in range(24):
    for j in range(60):
        if j == 0:
            date = str(i) + ":0" + str(j)
        else:
#             date = str(i) + ":" + str(j)
            date = ""
        temp.append(date)
datetime = temp[480:] + temp[:480]

plt.figure(figsize=(15,7), dpi=150)
plt.plot(data['MOR_1A'], linestyle='-.', label="MOR_1A", color="blue")
plt.plot(data['RVR_1A'], linestyle='-.', label="RVR_1A", color="red")
plt.plot(data['MOR_1A_cut'], linestyle='-.', label="MOR_1A_cut", color="green")

plt.xticks(np.arange(len(datetime)), datetime, rotation = 45)
plt.tick_params(labelsize=15)
plt.legend(fontsize=20)
plt.xlabel("Time",fontsize=25)
plt.ylabel("Visibility",fontsize=25)
plt.grid(which='major', axis='y')
plt.savefig('./display/'+ 'mor_rvr.jpg',bbox_inches='tight')
plt.show()
```
